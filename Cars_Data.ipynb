###  13.  What are the top 10 most represented car makes in the dataset? Are they focused on economy or premium cars?

```python
top_10_makes_list = data_cars['Make'].value_counts().head(10).index

avg_prices_top_10 = data_cars.groupby('Make')['MSRP'].mean().loc[top_10_makes_list]

avg_prices_top_10_sorted = avg_prices_top_10.sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=avg_prices_top_10_sorted.index, y=avg_prices_top_10_sorted.values)
plt.title('Average price (MSRP) of the 10 most popular brands')
plt.xlabel('Make')
plt.ylabel('Average Price (MSRP)')
plt.xticks(rotation=45)
plt.show()
```

### 14.Which cars offer the best compromise between: Affordable price (low MSRP), Good power (high Horsepower), Low fuel consumption (high MPG)

```python

normalized_df = data_cars.copy()
normalized_df['Horsepower_score'] = (normalized_df['Horsepower'] - normalized_df['Horsepower'].min()) / (normalized_df['Horsepower'].max() - normalized_df['Horsepower'].min())
normalized_df['MPG_City_score'] = (normalized_df['MPG_City'] - normalized_df['MPG_City'].min()) / (normalized_df['MPG_City'].max() - normalized_df['MPG_City'].min())

normalized_df['MSRP_score'] = 1 - (normalized_df['MSRP'] - normalized_df['MSRP'].min()) / (normalized_df['MSRP'].max() - normalized_df['MSRP'].min())

normalized_df['Compromis_Score'] = (normalized_df['MSRP_score'] + normalized_df['Horsepower_score'] + normalized_df['MPG_City_score']) / 3


plt.figure(figsize=(10, 6))

sns.scatterplot(x='MSRP', y='Horsepower', hue='Compromis_Score', style='Origin', size='Compromis_Score', data=normalized_df, palette='viridis', sizes=(20, 200))

plt.title('Price-Power-Consumption Compromise by Origin')
plt.xlabel('Price (MSRP)')
plt.ylabel('Horsepower')

plt.show()
```
